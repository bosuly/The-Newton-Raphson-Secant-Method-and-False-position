```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from typing import Callable, Optional

# ==========================================
# 1. SETUP & CONFIGURATION
# ==========================================
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# We store all 5 functions in a list of dictionaries for easy looping.
# logical_guesses helps us pick valid starting points for each specific function.
test_cases = [
    {
        "id": "A",
        "func": lambda x: x**5 + x - 1,
        "deriv": lambda x: 5*x**4 + 1,
        "formula": "x^5 + x - 1",
        "guess_nr": 0.5,       # Newton Initial Guess
        "bracket": (0, 1)      # Secant/False Position Bracket
    },
    {
        "id": "B",
        "func": lambda x: x**5 + 5*x**3 - 4*x + 1,
        "deriv": lambda x: 5*x**4 + 15*x**2 - 4,
        "formula": "x^5 + 5x^3 - 4x + 1",
        "guess_nr": 0.5,
        "bracket": (0, 1)
    },
    {
        "id": "C",
        "func": lambda x: x**5 + 2*x**4 - x - 3,
        "deriv": lambda x: 5*x**4 + 8*x**3 - 1,
        "formula": "x^5 + 2x^4 - x - 3",
        "guess_nr": 1.0,
        "bracket": (0, 2)
    },
    {
        "id": "D",
        "func": lambda x: x**5 - 10,
        "deriv": lambda x: 5*x**4,
        "formula": "x^5 - 10",
        "guess_nr": 1.5,       # x=0 would cause div by zero here!
        "bracket": (1, 2)
    },
    {
        "id": "E",
        "func": lambda x: x**5 + 2*x**2 + x - 0.5,
        "deriv": lambda x: 5*x**4 + 4*x + 1,
        "formula": "x^5 + 2x^2 + x - 0.5",
        "guess_nr": 0.5,
        "bracket": (0, 1)
    }
]

# ==========================================
# 2. ALGORITHM IMPLEMENTATIONS
# ==========================================

def newton_raphson(func, deriv, x0, tol=1e-6, max_iter=20):
    logs = []
    x_curr = x0
    
    for i in range(max_iter):
        f_val = func(x_curr)
        df_val = deriv(x_curr)
        
        if abs(df_val) < 1e-15:
            print(f"  [Error] Derivative zero at {x_curr}")
            break
            
        x_next = x_curr - (f_val / df_val)
        err = abs(x_next - x_curr)
        
        logs.append({"Iter": i+1, "x": x_curr, "f(x)": f_val, "Error": err})
        
        if err < tol:
            return x_next, pd.DataFrame(logs)
        x_curr = x_next
        
    return x_curr, pd.DataFrame(logs)

def secant_method(func, x0, x1, tol=1e-6, max_iter=20):
    logs = []
    
    for i in range(max_iter):
        f0 = func(x0)
        f1 = func(x1)
        
        if abs(f1 - f0) < 1e-15:
            print("  [Error] Division by zero in Secant")
            break
            
        x_new = x1 - f1 * (x1 - x0) / (f1 - f0)
        err = abs(x_new - x1)
        
        logs.append({"Iter": i+1, "x_prev": x0, "x_curr": x1, "Error": err})
        
        if err < tol:
            return x_new, pd.DataFrame(logs)
        x0, x1 = x1, x_new
        
    return x_new, pd.DataFrame(logs)

# ==========================================
# 3. VISUALIZATION FUNCTION
# ==========================================

def plot_function_analysis(case, root, x_range=(-2, 2)):
    x = np.linspace(x_range[0], x_range[1], 400)
    y = case["func"](x)
    
    plt.figure()
    plt.plot(x, y, label=f'f(x) = {case["formula"]}', linewidth=2, color='#2563eb')
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    
    # Mark the root
    if root is not None:
        plt.plot(root, case["func"](root), 'ro', label=f'Root: {root:.5f}', zorder=5)
        
    # Optional: Tangent Line at initial guess (for Newton visual)
    x0 = case["guess_nr"]
    y0 = case["func"](x0)
    m = case["deriv"](x0)
    tan_x = np.linspace(x0-1, x0+1, 100)
    tan_y = m*(tan_x - x0) + y0
    plt.plot(tan_x, tan_y, 'g--', alpha=0.6, label=f'Tangent at guess {x0}')

    plt.title(f"Analysis: {case['formula']}", fontsize=14)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

# ==========================================
# 4. MAIN LOOP (PROCESS ALL TASKS)
# ==========================================

for case in test_cases:
    print(f"\n{'='*60}")
    print(f" FUNCTION {case['id']}: {case['formula']}")
    print(f"{'='*60}")
    
    # --- A. Newton-Raphson ---
    print(f"1. Newton-Raphson (Guess: {case['guess_nr']})")
    root_nr, log_nr = newton_raphson(case["func"], case["deriv"], case["guess_nr"])
    print(log_nr.to_string(index=False)) # Print clean table
    print(f"   => Final Root: {root_nr:.6f}\n")
    
    # --- B. Secant Method ---
    print(f"2. Secant Method (Bracket: {case['bracket']})")
    root_sec, log_sec = secant_method(case["func"], case["bracket"][0], case["bracket"][1])
    print(log_sec.to_string(index=False))
    print(f"   => Final Root: {root_sec:.6f}\n")
    
    # --- C. Visualization ---
    # We use the root found by Newton-Raphson for the plot marker
    plot_function_analysis(case, root_nr, x_range=(-2.5, 2.5))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset

# ==========================================
# TASK 1: FUNCTION & DERIVATIVE DEFINITIONS
# ==========================================
# We define the mathematical functions and their calculus derivatives here.

def func_A(x): return x**5 + x - 1
def deriv_A(x): return 5*x**4 + 1

def func_B(x): return x**5 + 5*x**3 - 4*x + 1
def deriv_B(x): return 5*x**4 + 15*x**2 - 4

def func_C(x): return x**5 + 2*x**4 - x - 3
def deriv_C(x): return 5*x**4 + 8*x**3 - 1

def func_D(x): return x**5 - 10
def deriv_D(x): return 5*x**4

def func_E(x): return x**5 + 2*x**2 + x - 0.5
def deriv_E(x): return 5*x**4 + 4*x + 1

# Configuration Dictionary (Inputs for the algorithms)
test_cases = [
    {"id": "A", "func": func_A, "deriv": deriv_A, "str": "x^5 + x - 1", "x0": 0.5, "bracket": (0, 1)},
    {"id": "B", "func": func_B, "deriv": deriv_B, "str": "x^5 + 5x^3 - 4x + 1", "x0": 0.5, "bracket": (0, 1)},
    {"id": "C", "func": func_C, "deriv": deriv_C, "str": "x^5 + 2x^4 - x - 3", "x0": 1.0, "bracket": (0, 2)},
    {"id": "D", "func": func_D, "deriv": deriv_D, "str": "x^5 - 10", "x0": 1.5, "bracket": (1, 2)},
    {"id": "E", "func": func_E, "deriv": deriv_E, "str": "x^5 + 2x^2 + x - 0.5", "x0": 0.5, "bracket": (0, 1)},
]

# ==========================================
# TASK 2: ALGORITHM IMPLEMENTATION
# ==========================================

def newton_raphson(func, deriv, x0, tol=1e-6, max_iter=20):
    """
    Finds root using Newton-Raphson: x_new = x - f(x)/f'(x)
    """
    logs = []
    x_curr = x0
    
    for i in range(max_iter):
        f_val = func(x_curr)
        df_val = deriv(x_curr)
        
        # Safety Check: Derivative close to zero?
        if abs(df_val) < 1e-15:
            print(f"  [Error] Derivative too close to zero at x={x_curr:.4f}")
            break
            
        x_next = x_curr - (f_val / df_val)
        error = abs(x_next - x_curr)
        
        logs.append({
            "Iter": i+1, "x_curr": x_curr, "f(x)": f_val, 
            "f'(x)": df_val, "Error": error
        })
        
        if error < tol:
            return x_next, pd.DataFrame(logs)
        x_curr = x_next
        
    return x_curr, pd.DataFrame(logs)

def secant_method(func, x0, x1, tol=1e-6, max_iter=20):
    """
    Finds root using Secant Method (Approximated slope).
    """
    logs = []
    
    for i in range(max_iter):
        f0, f1 = func(x0), func(x1)
        
        # Safety Check: Division by zero?
        if abs(f1 - f0) < 1e-15:
            print("  [Error] Division by zero in Secant method.")
            break
            
        x_new = x1 - f1 * (x1 - x0) / (f1 - f0)
        error = abs(x_new - x1)
        
        logs.append({
            "Iter": i+1, "x_prev": x0, "x_curr": x1, 
            "x_new": x_new, "Error": error
        })
        
        if error < tol:
            return x_new, pd.DataFrame(logs)
        x0, x1 = x1, x_new
        
    return x_new, pd.DataFrame(logs)

# ==========================================
# TASK 3: VISUALIZATION (Plot + Tangent + Zoom)
# ==========================================

def plot_final_result(case, root):
    # Setup Data
    x_range = np.linspace(root - 2, root + 2, 400)
    y_vals = case["func"](x_range)
    
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # 1. Main Plot
    ax.plot(x_range, y_vals, label=f'f(x) = {case["str"]}', color='blue', linewidth=2)
    ax.axhline(0, color='black', linewidth=1, linestyle='--')
    ax.axvline(0, color='black', linewidth=1, linestyle='--')
    
    if root is not None:
        ax.plot(root, case["func"](root), 'ro', label=f'Root: {root:.5f}', zorder=5)

    # 2. Tangent Line Challenge
    # y - y0 = m(x - x0)
    x0 = case["x0"]
    y0 = case["func"](x0)
    m = case["deriv"](x0)
    tan_x = np.linspace(x0 - 1.5, x0 + 1.5, 100)
    tan_y = m * (tan_x - x0) + y0
    ax.plot(tan_x, tan_y, 'g--', alpha=0.6, label=f'Tangent at x={x0}')

    # 3. Zoomed Inset Feature
    if root is not None:
        axins = zoomed_inset_axes(ax, zoom=4, loc='lower right') 
        
        # Plot data on inset
        axins.plot(x_range, y_vals, 'b')
        axins.plot(root, case["func"](root), 'ro')
        axins.plot(tan_x, tan_y, 'g--')
        axins.axhline(0, color='black', linewidth=1, linestyle='--')
        
        # Smart Limits for Zoom Window
        x1, x2 = root - 0.2, root + 0.2
        y_zoom = case["func"](np.linspace(x1, x2, 50))
        y1, y2 = min(y_zoom), max(y_zoom)
        
        axins.set_xlim(x1, x2)
        axins.set_ylim(y1 - 0.1, y2 + 0.1)
        axins.set_xticks([]); axins.set_yticks([]) # Clean ticks
        axins.set_title("Zoom at Root")
        
        # Connect lines
        mark_inset(ax, axins, loc1=2, loc2=4, fc="none", ec="0.5")

    ax.set_title(f"Analysis: {case['str']}")
    ax.legend(loc='upper left')
    ax.grid(True, alpha=0.3)
    plt.show()

# ==========================================
# MAIN EXECUTION LOOP
# ==========================================
print("STARTING LAB BENCHMARK...\n")

for case in test_cases:
    print(f"\n{'='*60}")
    print(f" FUNCTION {case['id']}: {case['str']}")
    print(f"{'='*60}")
    
    # 1. Run Newton-Raphson
    print(f"1. Newton-Raphson (Guess: {case['x0']})")
    root_nr, log_nr = newton_raphson(case["func"], case["deriv"], case["x0"])
    print(log_nr.to_string(index=False))
    
    # 2. Run Secant
    print(f"\n2. Secant Method (Bracket: {case['bracket']})")
    root_sec, log_sec = secant_method(case["func"], case['bracket'][0], case['bracket'][1])
    print(log_sec.to_string(index=False))
    
    # 3. Generate Plot
    print(f"\n[Plotting {case['id']} with Zoom...]")
    plot_final_result(case, root_nr)
```
